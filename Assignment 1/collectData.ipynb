### Collect data from images using deep-face and save as a *.csv format.

```python
import os
from deepface import DeepFace
import pandas as pd
import numpy as np
```

```python
data = []
label = []
for index, dirname in enumerate(os.listdir('./data/')):
    for file in os.listdir(f'./data/{dirname}'):
        try:
            embedding = DeepFace.represent(img_path = f"data/{dirname}/{file}", model_name = 'ArcFace')    
            data.append(embedding)
            label.append(index)
        except:
            print(f"data/{dirname}/{file}")
```

```python
df = pd.DataFrame(data)
df['label'] = label
```

```python
df.to_csv('arcfaceData.csv', index=False)
```
